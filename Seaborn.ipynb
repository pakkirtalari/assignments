Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Scatter Plot: Used to show the relationship between two numerical variables. Useful for identifying correlations or patterns in data points.

Line Plot: Shows the relationship between two numerical variables connected by lines. Often used for time series data or sequential data.

Bar Plot: Used to compare values across different categories. Useful for showing frequency counts or aggregations.

Box Plot: Displays the distribution of a dataset's summary statistics, such as median, quartiles, and potential outliers. Useful for identifying the spread and skewness of data.

Pair Plot: Plots pairwise relationships in a dataset. Useful for quickly understanding the correlations between multiple numerical variables.

Q2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset("fmri")
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
plt.show()
```

Q3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic)
plt.title("Age by Pclass")

plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.title("Fare by Pclass")

plt.tight_layout()
plt.show()
```

Q4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset("diamonds")
sns.histplot(data=diamonds, x='price', hue='cut', multiple="stack")
plt.title("Price Histogram with Cut")
plt.show()
```

Q5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
sns.pairplot(data=iris, hue='species')
plt.show()
```

Q6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset("flights")
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(data=flights_pivot, cmap='YlGnBu', annot=True, fmt='d')
plt.title("Flights Heatmap")
plt.show()
```
