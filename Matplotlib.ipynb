Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a popular Python data visualization library used to create static, interactive, and animated visualizations in Python. It provides a wide range of plotting functionalities to help users create various types of charts, graphs, and plots.

Matplotlib is used for data visualization, which is the process of representing data in graphical form to understand patterns, relationships, and trends within the data. It's widely used in data analysis, scientific research, and presentations.

Five plots that can be plotted using the Pyplot module of Matplotlib are:

Line Plot

Scatter Plot

Bar Plot

Histogram

Pie Chart

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

A scatter plot is a type of data visualization that uses dots to represent the values obtained for two different variables. Each dot on the scatter plot represents a single data point. Scatter plots are used to analyze and visualize the relationship or correlation between two variables.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

The subplot() function in Matplotlib is used to create multiple plots within a single figure. It allows you to arrange multiple plots in rows and columns, making it useful for comparing different datasets or visualizations.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])
axes[0, 0].plot(x, y1)
axes[0, 0].set_title("Line 1")

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])
axes[0, 1].plot(x, y2)
axes[0, 1].set_title("Line 2")

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])
axes[1, 0].plot(x, y3)
axes[1, 0].set_title("Line 3")

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])
axes[1, 1].plot(x, y4)
axes[1, 1].set_title("Line 4")

plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot is a data visualization that uses rectangular bars to represent data values. It's often used to compare values between different categories. A bar plot can be vertical (showing bars going up) or horizontal (showing bars going sideways).

Bar plots are used to visually compare data points, identify trends, and analyze the distribution of values within different categories.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.title("Bar Plot")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()

# Horizontal Bar Plot
plt.barh(company, profit)
plt.title("Horizontal Bar Plot")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot (also known as a whisker plot) is a graphical representation of the distribution of a dataset. It displays the median, quartiles, and potential outliers in the data. It's used to identify the spread and skewness of the data.

Box plots are particularly useful for comparing distributions across different categories or groups and detecting outliers.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]
labels = ['Box 1', 'Box 2']

plt.boxplot(data, labels=labels)
plt.title("Box Plot")
plt.ylabel("Values")
plt.show()
```
