```python
import scipy.stats as stats

mean = 50
std_dev = 5
confidence_level = 0.95

sample_size = 100  # Choose an appropriate sample size

margin_of_error = stats.norm.ppf(1 - (1 - confidence_level) / 2) * (std_dev / (sample_size ** 0.5))
confidence_interval = (mean - margin_of_error, mean + margin_of_error)

print("95% Confidence Interval:", confidence_interval)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

df = 10  # Degrees of freedom
x = np.linspace(0, 30, 500)  # Values for x-axis
y = stats.chi2.pdf(x, df)  # Chi-square distribution PDF

plt.plot(x, y)
plt.xlabel('Chi-square Statistic')
plt.ylabel('Probability Density')
plt.fill_between(x[x <= 15], y[x <= 15], alpha=0.5)  # Shading the area
plt.title('Chi-square Distribution with 10 Degrees of Freedom')
plt.show()
```

```python
import scipy.stats as stats

observed_frequencies = [45, 55]  # Observed frequencies of tails and heads
expected_frequencies = [0.5, 0.5]  # Expected frequencies for a fair coin

chi2_statistic, p_value = stats.chisquare(observed_frequencies, expected_frequencies)

alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: The coin is biased.")
else:
    print("Fail to reject null hypothesis: The coin is fair.")
print("Chi-square Statistic:", chi2_statistic)
print("P-value:", p_value)
```

```python
import scipy.stats as stats

observed = [[60, 140], [30, 170]]  # Contingency table
chi2_statistic, p_value, dof, expected = stats.chi2_contingency(observed)

alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: There is a significant association between smoking status and lung cancer diagnosis.")
else:
    print("Fail to reject null hypothesis: There is no significant association.")
print("Chi-square Statistic:", chi2_statistic)
print("P-value:", p_value)
```
