```python

import seaborn as sns
import plotly.express as px

titanic = sns.load_dataset("titanic")
scatter_plot = px.scatter(titanic, x="age", y="fare", title="Scatter Plot of Age and Fare",color='sex')
scatter_plot.show()
```

```python
import plotly.express as px

tips = px.data.tips()
box_plot = px.box(tips, x="day", y="total_bill", title="Box Plot of Total Bill by Day")
box_plot.show()
```

```python
histogram = px.histogram(tips, x="sex", y="total_bill", color="day", pattern_shape="smoker",
                         title="Histogram of Total Bill by Sex and Day")
histogram.show()
```

```python
iris = px.data.iris()
scatter_matrix = px.scatter_matrix(iris, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
                                   color="species", title="Scatter Matrix Plot of Iris Dataset")
scatter_matrix.show()
```

A Distplot (Distribution Plot) is a type of plot that displays the distribution of a single variable. It combines a histogram with a kernel density estimate to provide an overall view of the probability distribution of the data.

```python
import plotly.express as px

tips = px.data.tips()
distplot = px.histogram(tips, x="total_bill", marginal="rug", title="Distplot of Total Bill")
distplot.show()
```
