Q1. What is the Probability density function?

A1. The Probability Density Function (PDF) is a statistical function that describes the likelihood of a continuous random variable taking on a specific value. In essence, the PDF provides the probability of a random variable falling within a particular range of values.

Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.

```python
import numpy as np
import scipy.stats

def normal_pdf(x, mean, std_dev):
    pdf_value = scipy.stats.norm.pdf(x, loc=mean, scale=std_dev)
    return pdf_value

mean = 0
std_dev = 1
x_value = 2
pdf = normal_pdf(x_value, mean, std_dev)
print("PDF at x =", x_value, ":", pdf)
```

Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.

A4. Properties of Binomial distribution:

The outcomes are binary (success/failure).
Each trial is independent.
The probability of success remains constant for each trial.
There is a fixed number of trials.
Examples of events where binomial distribution can be applied:

Flipping a coin multiple times and counting the number of heads.
Testing a product for defects, where each item can either be defective or non-defective.

Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plot a histogram of the results using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

sample_size = 1000
probability_success = 0.4
sample = np.random.binomial(n=1, p=probability_success, size=sample_size)

plt.hist(sample, bins=2, edgecolor='black')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.title('Histogram of Binomial Distribution')
plt.show()
```

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.

```python
import scipy.stats

def poisson_cdf(x, mean):
    cdf_value = scipy.stats.poisson.cdf(x, mu=mean)
    return cdf_value

mean = 5
x_value = 3
cdf = poisson_cdf(x_value, mean)
print("CDF at x =", x_value, ":", cdf)
```

Q7. How Binomial distribution different from Poisson distribution?

A7. The main differences between Binomial and Poisson distributions are:

Binomial distribution deals with a fixed number of independent trials, each with two possible outcomes (success/failure), while Poisson distribution deals with the number of events in a fixed interval.
Binomial distribution requires knowing the number of trials and the probability of success, whereas Poisson distribution requires knowing only the average rate of occurrence.

Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.

```python
import numpy as np

sample_size = 1000
mean = 5
sample = np.random.poisson(mean, size=sample_size)

sample_mean = np.mean(sample)
sample_variance = np.var(sample)

print("Sample Mean:", sample_mean)
print("Sample Variance:", sample_variance)
```

Q9. How mean and variance are related in Binomial distribution and Poisson distribution?

A9. In both Binomial and Poisson distributions, the mean and variance are related as follows:

For a Binomial distribution with parameters n (number of trials) and p (probability of success), mean = n * p and variance = n * p * (1 - p).
For a Poisson distribution, both the mean and variance are equal to the parameter λ (average rate of occurrence).

Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

A10. In a normal distribution, the least frequent data appears in the tails of the distribution, far away from the mean. This is because the normal distribution is symmetric, and data becomes progressively rarer as it moves away from the mean in both directions.
