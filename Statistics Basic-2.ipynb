Q1. What are the three measures of central tendency?

A1. The three measures of central tendency are:

Mean: The average of all values in a dataset.
Median: The middle value when the data is arranged in ascending or descending order.
Mode: The most frequently occurring value in a dataset.
Q2. What is the difference between the mean, median, and mode? How are they used to measure the central tendency of a dataset?

A2. The mean is the average value, calculated by summing all values and dividing by the number of values. The median is the middle value in the ordered dataset. The mode is the value that appears most often. They provide different ways to summarize the central value of data, with the median being less sensitive to outliers.

Q5. How are measures of dispersion such as range, variance, and standard deviation used to describe the spread of a dataset? Provide an example.

A5. Measures of dispersion quantify how much data points deviate from the central tendency. Range gives the interval between the minimum and maximum values. Variance and standard deviation measure the average squared difference from the mean. For example, a small standard deviation indicates data points are close to the mean, while a large one indicates more spread.

Q6. What is a Venn diagram?

A6. A Venn diagram is a graphical representation used to show the relationship between sets. It consists of overlapping circles that represent different sets, with shared and non-shared elements.

Q7. For the two given sets A = (2,3,4,5,6,7) & B = (0,2,6,8,10). Find:

(i) A ∩ B: The intersection of A and B, which are elements common to both sets. In this case, A ∩ B = (2,6).

(ii) A ⋃ B: The union of A and B, which are all elements present in either set. In this case, A ⋃ B = (0,2,3,4,5,6,7,8,10).

Q8. What do you understand about skewness in data?

A8. Skewness refers to the asymmetry of the distribution of data points around the mean. If the tail of the distribution is longer on one side, the data is skewed in that direction.

Q9. If a data is right skewed then what will be the position of median with respect to mean?

A9. If the data is right skewed, the tail of the distribution is longer on the right side (positive side). In this case, the median will be less than the mean.

Q10. Explain the difference between covariance and correlation. How are these measures used in statistical analysis?

A10. Covariance measures the extent to which two variables change together. Correlation measures both the strength and direction of the linear relationship between two variables. Covariance can take any value, while correlation is normalized between -1 and 1. These measures are used to understand the relationship between variables and to identify dependencies.

Q11. What is the formula for calculating the sample mean? Provide an example calculation for a dataset.

A11. The formula for calculating the sample mean (average) is:
Mean (x̄) = Σxi / n

Example:
Dataset: [12, 15, 18, 20, 25]
Mean = (12 + 15 + 18 + 20 + 25) / 5 = 18

Q12. For a normal distribution data what is the relationship between its measure of central tendency?

A12. For a normal distribution, the mean, median, and mode are equal and located at the center of the distribution.

Q13. How is covariance different from correlation?

A13. Covariance measures the degree to which two variables change together, while correlation measures both the strength and direction of the linear relationship between two variables. Covariance is not normalized, so its values can vary widely, while correlation is normalized to range between -1 and 1.

Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.

A14. Outliers can significantly affect the mean by pulling it towards extreme values. However, they have minimal impact on the median and mode. Outliers also affect measures of dispersion by increasing the range and standard deviation. For example, in a dataset of test scores, an extremely low score (outlier) can lower the mean score, even if most other scores are high.
