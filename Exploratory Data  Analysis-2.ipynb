```python
import seaborn as sns

# Load the flights dataset from Seaborn
data = sns.load_dataset("flights")

# Get the dimensions of the dataset
rows, columns = data.shape
print("Number of rows:", rows)
print("Number of columns:", columns)
data
```

```python
import matplotlib.pyplot as plt

# Create a histogram to visualize the distribution of flight prices
plt.figure(figsize=(10, 6))
sns.histplot(data['passengers'], bins=20, kde=True)
plt.xlabel('Number of Passengers')
plt.ylabel('Frequency')
plt.title('Distribution of Flight Passengers')
plt.show()

```

```python
min_price = data['passengers'].min()
max_price = data['passengers'].max()
price_range = max_price - min_price

print(f"Minimum Price: {min_price}")
print(f"Maximum Price: {max_price}")
print(f"Price Range: {price_range}")
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='year', y='passengers')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers by Year')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='year', y='passengers')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers by Year')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by month and calculate the mean number of passengers for each month
monthly_avg_passengers = data.groupby('month')['passengers'].mean()

# Create a line plot to show the trend of average passengers over the months
plt.figure(figsize=(10, 6))
sns.lineplot(x=monthly_avg_passengers.index, y=monthly_avg_passengers.values)
plt.xlabel('Month')
plt.ylabel('Average Number of Passengers')
plt.title('Average Number of Passengers by Month')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by year and calculate the mean number of passengers for each year
yearly_avg_passengers = data.groupby('year')['passengers'].mean()

# Create a line plot to show the trend of average passengers over the years
plt.figure(figsize=(10, 6))
sns.lineplot(x=yearly_avg_passengers.index, y=yearly_avg_passengers.values)
plt.xlabel('Year')
plt.ylabel('Average Number of Passengers')
plt.title('Average Number of Passengers by Year')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by year and month and calculate the mean number of passengers for each year and month combination
year_month_avg_passengers = data.groupby(['year', 'month'])['passengers'].mean().unstack()

# Create a heatmap to show how the average number of passengers varies across years and months
plt.figure(figsize=(12, 8))
sns.heatmap(year_month_avg_passengers, cmap='YlGnBu', annot=True, fmt=".0f", linewidths=.5)
plt.xlabel('Month')
plt.ylabel('Year')
plt.title('Average Number of Passengers by Year and Month')
plt.show()
```
